# Generate RGB imagery from level 2 PACE data and investigate associated Rrs spectra 

This code streams PACE OCI data using earthaccess to create RGB images from L2 surface reflectance products, as well as plot Rrs spectra from L2 AOP products, associated with user input target lat and lon co-ordinates. 

Code Updated: 12/11/2025

RGB imagery code is adpated from PACE Hackweek tutorials (https://pacehackweek.github.io/pace-2024/presentations/hackweek/satdata_visualization.html)

```python
# Load libraries
import numpy as np
import matplotlib.pyplot as plt
import xarray as xr
import cartopy
import cartopy.crs as ccrs
import earthaccess
```

```python
#log in to earth access
auth = earthaccess.login()
# are we authenticated?
if not auth.authenticated:
     #ask for credentials and persist them in a .netrc file
        auth.login(strategy="interactive", persist=True)
```

## First load Rhos product to create RGB imagery 

```python
#search for data 
# set lat/ lon mins and max (to create a bounding box for ROI)
lon_min = -31
lon_max = -22
lat_min = 51
lat_max = 63

tspan = ("2024-06-04", "2024-06-04") #date
bbox = (lon_min, lat_min, lon_max, lat_max) #lat lon
#clouds =  (0, 20) #percentage cloud cover 

results = earthaccess.search_data(
    short_name="PACE_OCI_L2_SFREFL",
    temporal=tspan,
    bounding_box=bbox,
)
results
```

```python
l2paths =earthaccess.open(results) #stream all results 
```

If you have more than one granule, choose the one you want for visualisation

In this example I know I want the 13:46 passover, which is granule number 6 (index 5 in python)


```python
#set up your data 
nav = xr.open_dataset(l2paths[5], group="navigation_data")
nav = nav.set_coords(("longitude","latitude"))
wave = xr.open_dataset(l2paths[5], group="sensor_band_parameters")
wave = wave.set_coords(("wavelength_3d"))
ds = xr.open_dataset(l2paths[5], group="geophysical_data")
dataset = xr.merge((ds["rhos"], nav, wave))
```

Then we subset the dataset to just include our ROI- this may take a little while but it will make future plotting etc a lot quicker

```python
# Subset to speed things up a bit (if you know where you want to look)
rhos_box = dataset.where((
        (dataset["latitude"] > lat_min)
        & (dataset["latitude"] < lat_max)
        & (dataset["longitude"] > lon_min)
        & (dataset["longitude"] < lon_max)),
    drop = True)
```

Here we select our lat/ lon points of interest

```python
#select lat lon points for data extraction 
spec_points = [(57.26, -25.13),
    (55.80, -25.71),
    (56.995,-29.176)]
```

Then we plot the true colour image with points of interest- this code is adapted from a PACE Hackweek tutorial.

```python
# This section includes adapted code from https://pacehackweek.github.io/pace-2024/presentations/hackweek/satdata_visualization.html

# OCI True Color 1 band -min/max adjusted
vmin = 0.01
vmax = 1.04 # Above 1 because whites can be higher than 1

rhos_red = rhos_box["rhos"].sel(wavelength_3d=645)  # Red channel at 645 nm
rhos_green = rhos_box["rhos"].sel(wavelength_3d=555)  # Green channel at 555 nm
rhos_blue = rhos_box["rhos"].sel(wavelength_3d=465)  # Blue channel at 465 nm

# Log stretch
red = np.log(rhos_red / 0.01) / np.log(1 / 0.01)
green = np.log(rhos_green / 0.01) / np.log(1 / 0.01)
blue = np.log(rhos_blue / 0.01) / np.log(1 / 0.01)

# Clip and stack
red = red.clip(min=vmin, max=vmax)
green = green.clip(min=vmin, max=vmax)
blue = blue.clip(min=vmin, max=vmax)

rgb = np.stack([red, green, blue], axis=-1)
rgb = (rgb - np.nanmin(rgb)) / (np.nanmax(rgb) - np.nanmin(rgb))

# Extract the longitude and latitude values for plotting
lon = rhos_box["longitude"].values
lat = rhos_box["latitude"].values

# Plot the true color image using Cartopy and Matplotlib
plt.figure(figsize=(10, 6))
ax = plt.axes(projection=cartopy.crs.PlateCarree())  # Set up the map projection
gl = ax.gridlines(draw_labels=True)  # Add gridlines with labels
gl.top_labels = False  # Hide top labels
gl.right_labels = False  # Hide right labels
gl.xlines = False  # Hide x-axis lines
gl.ylines = False  # Hide y-axis lines
gl.xlabel_style = {'size': 10}  # Customize label style
gl.ylabel_style = {'size': 10}  # Customize label style

# Plot match-up points
for i, (lat_pt, lon_pt) in enumerate(spec_points):
    ax.plot(lon_pt, lat_pt, 'o', color='yellow', markersize=6, markeredgecolor='black', transform=ccrs.PlateCarree())
    ax.text(lon_pt + 0.2, lat_pt + 0.2, f"P{i+1}", color='white', fontsize=9, weight='bold', transform=ccrs.PlateCarree())

# Add coastlines
ax.coastlines()

# Plot the RGB composite using pcolormesh and adjust brightness
plt.pcolormesh(lon, lat, np.clip(rgb * 1.5, 0, 1), shading='auto')  # brighten with *1.5
#plt.pcolormesh(lon, lat, rgb, shading='auto') #non-brightening plot
plt.xlabel("Longitude")  # Label the x-axis
plt.ylabel("Latitude")   # Label the y-axis
plt.title("RGB Composite (640, 555, 465 nm)")  # Title the plot
plt.xlim(lon_min, lon_max)  # Set longitude limits (cropping the region)
plt.ylim(lat_min, lat_max)    # Set latitude limits (cropping the region)
plt.show()  # Display the plot
```

## Load in Rrs data to plot Rrs at chosen points 

With PACE, the L2 surface reflectance product and the L2 AOP product are both available as standard from earth data, which is useful. Previously, if you wanted to create true colour imagery you would have to process data from L1A yourself and include the rhos in l2gen. However, we now have to make sure we are downloading the Rrs granule that corresponds to the data we plotted above (you can do this by checking the time stamps). This ensures your Rrs spectra will correspond to the imagery plotted. 

```python
rrs_results = earthaccess.search_data(
    short_name="PACE_OCI_L2_AOP",
    temporal=tspan,
    bounding_box=bbox,
)
rrs_results
```

```python
l2_rrs =earthaccess.open(rrs_results) #stream all results 
```

You need to make sure your timestamp matches the timestamp of the surface reflectance product. Change the index of l2_rrs to match the correct timestamp.

n/b this might not be the same index as the surface reflectance product so make sure to check. 

```python
#setup Rrs dataset 
rrs_nav = xr.open_dataset(l2_rrs[5], group="navigation_data")
rrs_nav = rrs_nav.set_coords(("longitude","latitude"))
rrs_wave = xr.open_dataset(l2_rrs[5], group="sensor_band_parameters")
rrs_wave = rrs_wave.set_coords(("wavelength_3d"))
rrs_ds = xr.open_dataset(l2_rrs[5], group="geophysical_data")
rrs_dataset = xr.merge((rrs_ds["Rrs"], rrs_nav, rrs_wave))
wavelengths = rrs_dataset['wavelength_3d'].values
```

Here we select a 3 by 3 pixel radius around our points of interest. This follows NASA's standard matchup procedure.

```python
def get_3x3_spectra(rrs_da, lat_arr, lon_arr, target_lat, target_lon):
    """
    Returns an array of Rrs spectra for pixels around the target lat/lon.
    Uses a 3x3 pixel neighborhood around the nearest pixel.
    Distance is computed using Plate Carrée projection (longitude scaled by cos(latitude)).
    """

    # Flatten lat, lon, and Rrs
    lat_flat = lat_arr.values.flatten()
    lon_flat = lon_arr.values.flatten()
    rrs_flat = rrs_da.values.reshape(-1, rrs_da.sizes['wavelength_3d'])

    # Approximate Plate Carrée distance in degrees
    dlat = lat_flat - target_lat
    # Scale longitude by cos(latitude) to account for convergence of meridians
    dlon = (lon_flat - target_lon) * np.cos(np.deg2rad(target_lat))
    distances = np.sqrt(dlat**2 + dlon**2)

    # Find the nearest pixel
    center_idx = distances.argmin()

    # Convert flattened index back to 2D indices
    n_lines, n_pixels = lat_arr.shape
    line_idx = center_idx // n_pixels
    pixel_idx = center_idx % n_pixels

    # 3x3 neighborhood slices
    line_slice = slice(max(line_idx-1, 0), min(line_idx+2, n_lines))
    pixel_slice = slice(max(pixel_idx-1, 0), min(pixel_idx+2, n_pixels))

    # Boolean mask for neighborhood
    neighborhood_mask = np.zeros(lat_flat.shape, dtype=bool)
    for i in range(line_slice.start, line_slice.stop):
        for j in range(pixel_slice.start, pixel_slice.stop):
            idx = i * n_pixels + j
            neighborhood_mask[idx] = True

    # Extract spectra, remove NaNs
    neighborhood_pixels = rrs_flat[neighborhood_mask, :]
    neighborhood_pixels = neighborhood_pixels[~np.isnan(neighborhood_pixels).any(axis=1)]

    return neighborhood_pixels
```

```python
#plot spectra 
plt.figure(figsize=(12, 6))

for i, (lat_pt, lon_pt) in enumerate(spec_points, start=1):
    spectra_pixels = get_3x3_spectra(
        rrs_dataset['Rrs'], rrs_dataset['latitude'], rrs_dataset['longitude'],
        lat_pt, lon_pt
    )

    if spectra_pixels.shape[0] == 0:
        print(f"No valid pixels around ({lat_pt}, {lon_pt})")
        continue

    mean_spectrum = np.nanmean(spectra_pixels, axis=0)
    std_spectrum = np.nanstd(spectra_pixels, axis=0)

    plt.plot(wavelengths, mean_spectrum, marker='o', label=f"Point {i}")
    plt.fill_between(
        wavelengths,
        mean_spectrum - std_spectrum,
        mean_spectrum + std_spectrum,
        alpha=0.2
    )

plt.title("Mean PACE OCI Rrs Spectra (3x3 pixels) around chosen points")
plt.xlabel("Wavelength (nm)")
plt.ylabel("Rrs (sr⁻¹)")
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

### Sense check to make sure the Rrs spectra are coming from the right place 

This is not a necessary step I just like to check my Rrs spectra are coming from the right place 

Also if you get a no data error in the above code you can check to see if there's actually no data 

```python
# crop rrs data to same extent as rhos data 
rrs_box = rrs_dataset.where(
    (rrs_nav["latitude"] > lat_min) &
    (rrs_nav["latitude"] < lat_max) &
    (rrs_nav["longitude"] > lon_min) &
    (rrs_nav["longitude"] < lon_max),
    drop=True
)
# Get lat/lon for this box
lats = rrs_box["latitude"].values
lons = rrs_box["longitude"].values

#choose wl for plotting
target_wavelength = 550
w_idx = np.abs(rrs_wave["wavelength_3d"].values - target_wavelength).argmin()
rrs_layer = rrs_box["Rrs"].isel(wavelength_3d=w_idx)
```

```python
#plot rrs with points overlayed
plt.figure(figsize=(10, 6))

# Create map axes with PlateCarree projection
ax = plt.axes(projection=ccrs.PlateCarree())

# Plot the Rrs data using pcolormesh
mesh = ax.pcolormesh(
    lons, lats, rrs_layer.values,
    cmap='viridis',
    shading='auto'
)

# Add colorbar
cbar = plt.colorbar(mesh, ax=ax, orientation='vertical', pad=0.02)
cbar.set_label(f"Rrs at {rrs_wave['wavelength_3d'].values[w_idx]:.0f} nm")

# Overlay target points
spec_points_in_box = [
    (lat, lon) for (lat, lon) in spec_points
    if lat_min < lat < lat_max and lon_min < lon < lon_max
]
if spec_points_in_box:
    spec_lats, spec_lons = zip(*spec_points_in_box)
    ax.scatter(spec_lons, spec_lats, color='red', marker='x', s=50, transform=ccrs.PlateCarree(), label="Target points")
    
# Add 3×3 bounding boxes for each target point
    n_lines, n_pixels = lats.shape
   

# Add coastlines and gridlines
ax.coastlines(resolution='10m')
gl = ax.gridlines(draw_labels=True)  # Add gridlines with labels
gl.top_labels = False  # Hide top labels
gl.right_labels = False  # Hide right labels
plt.xlim(lon_min, lon_max)  # Set longitude limits (cropping the region)
plt.ylim(lat_min, lat_max) 
plt.title("Rrs with target points for sense check")
plt.legend()
plt.show()
```

## Looking at standard chlorophyll a product
I also want to look at the standard chl product to see what the concs are like in the cocco bloom- for this i need level 2 bgc data (biogeochemical)

```python
bgc_results = earthaccess.search_data(
    short_name="PACE_OCI_L2_BGC",
    temporal=tspan,
    bounding_box=bbox,
)
bgc_results
```

```python
l2_bgc =earthaccess.open(bgc_results) #stream all results 
```

```python
#setup Rrs dataset 
bgc_nav = xr.open_dataset(l2_bgc[5], group="navigation_data")
bgc_nav = bgc_nav.set_coords(("longitude","latitude"))
bgc_ds = xr.open_dataset(l2_bgc[5], group="geophysical_data")
chl_dataset = xr.merge((bgc_ds["chlor_a"], bgc_nav))
```

```python
import cartopy.feature as cfeature
# Extract data
lon = chl_dataset['longitude']
lat = chl_dataset['latitude']
chl = chl_dataset['chlor_a']

# Define min and max for colorbar
vmin = 0.0   # minimum chlorophyll-a
vmax = 5.0  # maximum chlorophyll-a

# Create figure and map projection
fig, ax = plt.subplots(figsize=(10,6), subplot_kw={'projection': ccrs.PlateCarree()})

# Plot chlorophyll as pcolormesh
pcm = ax.pcolormesh(lon, lat, chl, cmap='viridis', shading='auto',vmin=vmin,vmax=vmax)

# Add coastlines and features
ax.coastlines()
ax.add_feature(cfeature.LAND)
ax.add_feature(cfeature.BORDERS, linestyle=':')

# Add colorbar
cbar = plt.colorbar(pcm, ax=ax, orientation='vertical', label='Chlorophyll-a (mg m^-3)')

# Add title
ax.set_title('PACE standard Chlorophyll-a product')
gl = ax.gridlines(draw_labels=True,alpha=0.3)  # Add gridlines with labels
gl.top_labels = False  # Hide top labels
gl.right_labels = False  # Hide right labels
plt.xlim(lon_min, lon_max)  # Set longitude limits (cropping the region)
plt.ylim(lat_min, lat_max) 
plt.show()
```
